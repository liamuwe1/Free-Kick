```python
def derivX(u):
    return u
def derivU(u,v,w):
    return -(rho*A*C_d)/(2*m)*(u**2+v**2+w**2)**0.5*u + rho*A*s*np.cos(alpha)*C_l*(u**2+v**2+w**2)/(2*m*(u**2+w**2)**0.5)*-np.abs(w) + rho*A*s*np.sin(alpha)*C_l*(u**2+v**2+w**2)**0.5/(2*m*(u**2*v**2+v**2*w**2+(u**2+w**2)**2)**0.5)*u*v

def derivY(v):
    return v
def derivV(u,v,w):
    return -(rho*A*C_d)/(2*m)*(u**2+v**2+w**2)**0.5*v + rho*A*s*np.sin(alpha)*C_l*(u**2+v**2+w**2)**0.5/(2*m*(u**2*v**2+v**2*w**2+(u**2+w**2)**2)**0.5)*-(u**2+v**2)

def derivZ(w):
    return w
def derivW(u,v,w):
    return -g -(rho*A*C_d)/(2*m)*(u**2+v**2+w**2)**0.5*w + rho*A*s*np.cos(alpha)*C_l*(u**2+v**2+w**2)/(2*m*(u**2+w**2)**0.5)*u + rho*A*s*np.sin(alpha)*C_l*(u**2+v**2+w**2)**0.5/(2*m*(u**2*v**2+v**2*w**2+(u**2+w**2)**2)**0.5)*v*w
```

```python
def RK4(fun1, fun2, fun3, fun4, fun5, fun6, h, t0, x0, y0, z0, u0, v0, w0, Nsteps):
    #RK4
    tt = [t0]
    xx = [x0]
    uu = [u0]
    yy = [y0]
    vv = [v0]
    zz = [z0] 
    ww = [w0]
    
    for i in range (Nsteps):
        
        K1x, K1y, K1z = fun1(uu[i]), fun3(vv[i]), fun5(ww[i])
        K1u, K1v, K1w = fun2(uu[i], vv[i], ww[i]), fun4(uu[i], vv[i], ww[i]), fun6(uu[i], vv[i], ww[i])
        
        K2x, K2y, K2z = fun1(uu[i] + h * K1x / 2), fun3(vv[i] + h * K1y / 2), fun5(ww[i] + h * K1z / 2)
        K2u, K2v, K2w = fun2(uu[i] + h * K1u / 2, vv[i] + h * K1v / 2, ww[i] + h * K1w / 2), fun4(uu[i] + h * K1u / 2, vv[i] + h * K1v / 2, ww[i] + h * K1w / 2), fun6(uu[i] + h * K1u / 2, vv[i] + h * K1v / 2, ww[i] + h * K1w / 2)
        
        K3x, K3y, K3z = fun1(uu[i] + h * K2x / 2), fun3(vv[i] + h * K2y / 2), fun5(ww[i] + h * K2z / 2)
        K3u, K3v, K3w = fun2(uu[i] + h * K2u / 2, vv[i] + h * K2v / 2, ww[i] + h * K2w / 2), fun4(uu[i] + h * K2u / 2, vv[i] + h * K2v / 2, ww[i] + h * K2w / 2), fun6(uu[i] + h * K2u / 2, vv[i] + h * K2v / 2, ww[i] + h * K2w / 2)
        
        K4x, K4y, K4z = fun1(uu[i] + h * K3x), fun3(vv[i] + h * K3y), fun5(ww[i] + h * K3z)
        K4u, K4v, K4w = fun2(uu[i] + h * K3u, vv[i] + h * K3v, ww[i] + h * K3w), fun4(uu[i] + h * K3u, vv[i] + h * K3v, ww[i] + h * K3w), fun6(uu[i] + h * K3u, vv[i] + h * K3v, ww[i] + h * K3w)
        
        xnext = xx[i] + (h/6) * (K1x + 2 * K2x + 2 * K3x + K4x)
        unext = uu[i] + (h/6) * (K1u + 2 * K2u + 2 * K3u + K4u)
        ynext = yy[i] + (h/6) * (K1y + 2 * K2y + 2 * K3y + K4y)
        vnext = vv[i] + (h/6) * (K1v + 2 * K2v + 2 * K3v + K4v)
        znext = zz[i] + (h/6) * (K1z + 2 * K2z + 2 * K3z + K4z)
        wnext = ww[i] + (h/6) * (K1w + 2 * K2w + 2 * K3w + K4w)
        
        tt.append(tt[i] + h)
        xx.append(xnext)
        uu.append(unext)
        yy.append(ynext)
        vv.append(vnext)
        zz.append(znext)
        ww.append(wnext)        
        
        sigma = np.arctan((abs(y0)-3.65)/x0)
        tau = np.arctan(ws/9.1)
        B = (9.1**2+ws**2)**0.5

        if y0 > 3.65 and (x0-B*np.cos(sigma+tau)) > xx[i+1] > (x0-9.1*np.cos(sigma)) and (abs(y0)-B*np.sin(sigma+tau)) < yy[i+1] < (abs(y0)-9.1*np.sin(sigma)) and 0 < zz[i+1] < 1.8:
            break
            
        if y0 < -3.65 and (x0-B*np.cos(sigma+tau)) > xx[i+1] > (x0-9.1*np.cos(sigma)) and -(abs(y0)-B*np.sin(sigma+tau)) > yy[i+1] > -(abs(y0)-9.1*np.sin(sigma)) and 0 < zz[i+1] < 1.8:
            break
            
        if 0 < y0 < 3.65 and (x0-B*np.cos(sigma+tau)) < xx[i+1] < (x0-9.1*np.cos(sigma)) and (abs(y0)-B*np.sin(sigma+tau)) < yy[i+1] < (abs(y0)-9.1*np.sin(sigma)) and 0 < zz[i+1] < 1.8:
            break
            
        if 0 > y0 > -3.65 and (x0-B*np.cos(sigma+tau)) < xx[i+1] < (x0-9.1*np.cos(sigma)) and -(abs(y0)-B*np.sin(sigma+tau)) > yy[i+1] > -(abs(y0)-9.1*np.sin(sigma)) and 0 < zz[i+1] < 1.8:
            break
            
        if xx[i+1] < -0.11 or xx[i+1] > 100.11 or yy[i+1] < -30.11 or yy[i+1] > 30.11 or zz[i+1] < 0:
            break
    
    #Figure details
    fig = plt.figure(figsize=(9,6))
    plt.rcParams['font.size'] = '12'
    plt.rcParams['font.sans-serif'] = ['Times New Roman']
    ax = plt.axes(projection='3d')
    ax.scatter3D(x0, y0, z0, s=15, c='k')
    ax.axes.set_xlim3d(-5, x0+5) 
    ax.axes.set_ylim3d(-20, 20) 
    ax.axes.set_zlim3d(0, 10) 
    scale=np.diag([x0+10, 40, 10, 1.0])
    scale=scale*(1.0/scale.max())
    scale[3,3]=1.0
    def short_proj():
      return np.dot(Axes3D.get_proj(ax), scale)
    ax.get_proj=short_proj
    fig.gca().yaxis.set_major_locator(mticker.MultipleLocator(2))
    fig.gca().zaxis.set_major_locator(mticker.MultipleLocator(2))
    ax.grid(False)

    #results
    
    if not xx[i+1] < -0.11 and not xx[i+1] > 100.11 and not yy[i+1] < -30.11 and not yy[i+1] > 30.11 and not zz[i+1] < 0:
        ax.text(0, 4, 4, "Hit the wall", color='red')
        label = np.around(xx[i+1],2), np.around(yy[i+1],2), np.around(zz[i+1],2)
        label2 = np.around(tt[i+1],2)
        ax.text(0, 4, 8, label, color='black')
        ax.text(0, 4, 6, label2, color='black')
    
    if zz[i] <  0.11:
        ax.text(0, 4, 4, "Didn't reach the goal", color='red')
        label = np.around(xx[i+1],2), np.around(yy[i+1],2), np.around(zz[i+1],2)
        label2 = np.around(tt[i+1],2)
        ax.text(0, 4, 8, label, color='black')
        ax.text(0, 4, 6, label2, color='black')
        
    if -2.65 < yy[i+1] < 2.65 and 0.11 < zz[i+1] < 2.44 and xx[i+1] < -0.11:
        ax.text(0, 4, 4, "Saved", color='orange')
        label = np.around(xx[i+1],2), np.around(yy[i+1],2), np.around(zz[i+1],2)
        label2 = np.around(tt[i+1],2)
        ax.text(0, 4, 8, label, color='black')
        ax.text(0, 4, 6, label2, color='black')
        
    if (2.65 < yy[i+1] < 3.65 or -2.65 > yy[i+1] > -3.65) and 0.11 < zz[i+1] < 2.44 and xx[i+1] < -0.11 and tt[i+1] < 1:
        ax.text(0, 4, 4, "Goal", color='green')
        label = np.around(xx[i+1],2), np.around(yy[i+1],2), np.around(zz[i+1],2)
        label2 = np.around(tt[i+1],2)
        ax.text(0, 4, 8, label, color='black')
        ax.text(0, 4, 6, label2, color='black')
        
    if (2.65 < yy[i+1] < 3.65 or -2.65 > yy[i+1] > -3.65) and 0.11 < zz[i+1] < 2.44 and xx[i+1] < -0.11 and tt[i+1] >= 1:
        ax.text(0, 4, 4, "Goal", color='green')
        label = np.around(xx[i+1],2), np.around(yy[i+1],2), np.around(zz[i+1],2)
        label2 = np.around(tt[i+1],2)
        ax.text(0, 4, 8, label, color='black')
        ax.text(0, 4, 6, label2, color='black')    
         
    if (abs(yy[i+1]) > 3.65 or zz[i+1] > 2.44) and xx[i+1] < -0.11:
        ax.text(0, 4, 4, "Miss", color='red')
        label = np.around(xx[i+1],2), np.around(yy[i+1],2), np.around(zz[i+1],2)
        label2 = np.around(tt[i+1],2)
        ax.text(0, 4, 8, label, color='black')
        ax.text(0, 4, 6, label2, color='black')    
    
    #Football pitch
    l=1
    
    p1 = [0,-30,0]
    p2 = [100,-30,0]
    p3 = [100,30,0]
    p4=[0,30,0]

    x_pitch=[p1[0],p2[0],p3[0],p4[0],p1[0]]
    y_pitch=[p1[1],p2[1],p3[1],p4[1],p1[1]]
    z_pitch=[p1[2],p2[2],p3[2],p4[2],p1[2]]

    ax.plot3D(x_pitch, y_pitch, z_pitch, '-w',linewidth=l)

    p1=[0,-20,0]
    p2=[16.5,-20,0]
    p3=[16.5,20,0]
    p4=[0,20,0]

    x_box=[p1[0],p2[0],p3[0],p4[0],p1[0]]
    y_box=[p1[1],p2[1],p3[1],p4[1],p1[1]]
    z_box=[p1[2],p2[2],p3[2],p4[2],p1[2]]

    ax.plot3D(x_box,y_box,z_box, '-w',linewidth=l)

    p1=[100,-20,0]
    p2=[83.5,-20,0]
    p3=[83.5,20,0]
    p4=[100,20,0]

    x_boxo=[p1[0],p2[0],p3[0],p4[0],p1[0]]
    y_boxo=[p1[1],p2[1],p3[1],p4[1],p1[1]]
    z_boxo=[p1[2],p2[2],p3[2],p4[2],p1[2]]

    ax.plot3D(x_boxo,y_boxo,z_boxo, '-w',linewidth=l)

    p1=[0,-9,0]
    p2=[5.5,-9,0]
    p3=[5.5,9,0]
    p4=[0,9,0]

    x_6box=[p1[0],p2[0],p3[0],p4[0],p1[0]]
    y_6box=[p1[1],p2[1],p3[1],p4[1],p1[1]]
    z_6box=[p1[2],p2[2],p3[2],p4[2],p1[2]]

    ax.plot3D(x_6box,y_6box,z_6box, '-w',linewidth=l)

    p1=[100,-9,0]
    p2=[94.5,-9,0]
    p3=[94.5,9,0]
    p4=[100,9,0]

    x_6boxo=[p1[0],p2[0],p3[0],p4[0],p1[0]]
    y_6boxo=[p1[1],p2[1],p3[1],p4[1],p1[1]]
    z_6boxo=[p1[2],p2[2],p3[2],p4[2],p1[2]]

    ax.plot3D(x_6boxo,y_6boxo,z_6boxo, '-w',linewidth=l)

    p1=[50,-30,0]
    p2=[50,30,0]

    x_centreline=[p1[0],p2[0]]
    y_centreline=[p1[1],p2[1]]
    z_centreline=[p1[2],p2[2]]

    ax.plot3D(x_centreline,y_centreline,z_centreline, '-w',linewidth=l)

    p1=[0,-3.65,0]
    p2=[0,-3.65,2.44]
    p3=[0,3.65,2.44]
    p4=[0,3.65,0]

    x_goal=[p1[0],p2[0],p3[0],p4[0],p1[0]]
    y_goal=[p1[1],p2[1],p3[1],p4[1],p1[1]]
    z_goal=[p1[2],p2[2],p3[2],p4[2],p1[2]]

    ax.plot3D(x_goal,y_goal,z_goal, '-w',linewidth=l)

    p1=[100,-3.65,0]
    p2=[100,-3.65,2.44]
    p3=[100,3.65,2.44]
    p4=[100,3.65,0]

    x_goalo=[p1[0],p2[0],p3[0],p4[0],p1[0]]
    y_goalo=[p1[1],p2[1],p3[1],p4[1],p1[1]]
    z_goalo=[p1[2],p2[2],p3[2],p4[2],p1[2]]

    ax.plot3D(x_goalo,y_goalo,z_goalo, '-w',linewidth=l)

    p1=[11,0,0]

    x_pspot=[p1[0]]
    y_pspot=[p1[1]]
    z_pspot=[p1[2]]

    ax.plot3D(x_pspot,y_pspot,z_pspot, '.w', markersize=2)

    p1=[89,0,0]

    x_pspoto=[p1[0]]
    y_pspoto=[p1[1]]
    z_pspoto=[p1[2]]

    ax.plot3D(x_pspoto,y_pspoto,z_pspoto, '.w', markersize=2)

    p1=[0,-3.65,0]
    p2=[-1,-3.65,0]
    p3=[0,-3.65,2.44]

    x_lstrut=[p1[0],p2[0],p3[0]]
    y_lstrut=[p1[1],p2[1],p3[1]]
    z_lstrut=[p1[2],p2[2],p3[2]]

    #ax.plot3D(x_lstrut,y_lstrut,z_lstrut,'-w',linewidth=l)

    p1=[0,3.65,0]
    p2=[-1,3.65,0]
    p3=[0,3.65,2.44]

    x_rstrut=[p1[0],p2[0],p3[0]]
    y_rstrut=[p1[1],p2[1],p3[1]]
    z_rstrut=[p1[2],p2[2],p3[2]]

    #ax.plot3D(x_rstrut,y_rstrut,z_rstrut,'-w',linewidth=l)

    p1=[-1,-3.65,0]
    p2=[-1,3.65,0]

    x_bstrut=[p1[0],p2[0]]
    y_bstrut=[p1[1],p2[1]]
    z_bstrut=[p1[2],p2[2]]

    #ax.plot3D(x_bstrut,y_bstrut,z_bstrut,'-w',linewidth=l)

    p1=[100,-3.65,0]
    p2=[101,-3.65,0]
    p3=[100,-3.65,2.44]

    x_lstruto=[p1[0],p2[0],p3[0]]
    y_lstruto=[p1[1],p2[1],p3[1]]
    z_lstruto=[p1[2],p2[2],p3[2]]

    #ax.plot3D(x_lstruto,y_lstruto,z_lstruto,'-w',linewidth=l)

    p1=[100,3.65,0]
    p2=[101,3.65,0]
    p3=[100,3.65,2.44]

    x_rstruto=[p1[0],p2[0],p3[0]]
    y_rstruto=[p1[1],p2[1],p3[1]]
    z_rstruto=[p1[2],p2[2],p3[2]]

    #ax.plot3D(x_rstruto,y_rstruto,z_rstruto,'-w',linewidth=l)

    p1=[101,-3.65,0]
    p2=[101,3.65,0]

    x_bstruto=[p1[0],p2[0]]
    y_bstruto=[p1[1],p2[1]]
    z_bstruto=[p1[2],p2[2]]

    #ax.plot3D(x_bstruto,y_bstruto,z_bstruto,'-w',linewidth=l)
    
    arcarc = np.linspace(-53, 53, 100)

    xxArc = 11+9.15*np.cos(arcarc*np.pi/180)    
    yyArc = 9.15*np.sin(arcarc*np.pi/180)    
    zzArc = 0

    ax.plot3D(xxArc, yyArc, zzArc, 'w',linewidth=l)

    xxArco = 89-9.15*np.cos(arcarc*np.pi/180)    
    yyArco = -9.15*np.sin(arcarc*np.pi/180)    
    zzArco = 0

    ax.plot3D(xxArco, yyArco, zzArco, 'w',linewidth=l)

    circir = np.linspace(0,np.pi*2,100)

    x_semcir = 50+9.15*np.cos(circir)
    y_semcir = 9.15*np.sin(circir)
    z_semcir = 0

    ax.plot3D(x_semcir,y_semcir,z_semcir, '-w',linewidth=l)
    
    
    fig.set_size_inches(13, 13)
    ax.w_zaxis.set_pane_color((0,0.34,0, 0.7))
    plt.xticks([])
    plt.yticks([])
    ax.set_zticks([])
    
    xlp = [xx[i], xx[i-1], xx[i-2], xx[i-3]]
    ylp = [yy[i], yy[i-1], yy[i-2], yy[i-3]]
    zlp = [zz[i], zz[i-1], zz[i-2], zz[i-3]]
    
    #Wall    
    if y0 < 0:
        p1=[x0-9.1*np.cos(sigma), (-1)*(abs(y0)-9.1*np.sin(sigma)), 0]
        p2=[x0-9.1*np.cos(sigma), (-1)*(abs(y0)-9.1*np.sin(sigma)), 1.8]
        p3=[x0-B*np.cos(sigma+tau), (-1)*(abs(y0)-B*np.sin(sigma+tau)), 1.8]
        p4=[x0-B*np.cos(sigma+tau), (-1)*(abs(y0)-B*np.sin(sigma+tau)), 0]
    
        x_wall=[p1[0],p2[0],p3[0],p4[0],p1[0]]
        y_wall=[p1[1],p2[1],p3[1],p4[1],p1[1]]
        z_wall=[p1[2],p2[2],p3[2],p4[2],p1[2]]
        
        wall = [list(zip(x_wall, y_wall, z_wall))]
        ax.add_collection3d(Poly3DCollection(wall, edgecolors='k', facecolors='k'))        
        
    if y0 >= 0:
        p1=[x0-9.1*np.cos(sigma), (abs(y0)-9.1*np.sin(sigma)), 0]
        p2=[x0-9.1*np.cos(sigma), (abs(y0)-9.1*np.sin(sigma)), 1.8]
        p3=[x0-B*np.cos(sigma+tau), (abs(y0)-B*np.sin(sigma+tau)), 1.8]
        p4=[x0-B*np.cos(sigma+tau), (abs(y0)-B*np.sin(sigma+tau)), 0]
        
        x_wall=[p1[0],p2[0],p3[0],p4[0],p1[0]]
        y_wall=[p1[1],p2[1],p3[1],p4[1],p1[1]]
        z_wall=[p1[2],p2[2],p3[2],p4[2],p1[2]]
        
        wall = [list(zip(x_wall, y_wall, z_wall))]
        ax.add_collection3d(Poly3DCollection(wall, edgecolors='k', facecolors='k')) 
    
    ax.plot3D(xx, yy, zz, c='r')
    ax.scatter3D(xx[i], yy[i], zz[i], s=15, c='k')
    
    p1=[0,-3.53,2.41]
    p2=[0,-2.67,2.4]
    p3=[0,-2.68,2.28]
    p4=[0,-3.53,2.32]

    x_goalarea=[p1[0],p2[0],p3[0], p4[0]]
    y_goalarea=[p1[1],p2[1],p3[1], p4[1]]
    z_goalarea=[p1[2],p2[2],p3[2], p4[2]]
    
    goalarea = [list(zip(x_goalarea, y_goalarea, z_goalarea))]
    #ax.add_collection3d(Poly3DCollection(goalarea, edgecolors='k', facecolors='w'))
```
